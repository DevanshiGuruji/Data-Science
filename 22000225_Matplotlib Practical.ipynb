```python
import numpy as np 
x=np.linspace(0,5,11)
y=x
```

```python
plt.plot(x,y,'b*-')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.title('This is a graph')
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'c*-')
plt.xlabel('x axis of subplot1')
plt.ylabel('y axis of subplot1')



plt.subplot(1,2,2)
plt.plot(x,y,'r<-')
plt.xlabel('x axis of subplot2')
plt.ylabel('y axis of subplot2')
```

Introduction to the object oriented method

```python
fig, axes = plt.subplots(2,2)

axes[0,0].plot(x,y,'r<-')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')
axes[0,0].set_title('title');

axes[0,1].plot(x,y,'c*-')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')
axes[0,1].set_title('title')

axes[1,0].plot(x,y,'y^-')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')
axes[1,0].set_title('title')

axes[1,1].plot(x,y,'g+-')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')
axes[1,1].set_title('title')

fig.savefig('plot.png')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.6,0.3,0.2])

axes1.plot(x,y,'c*-')
axes1.set_xlabel('x_label_axes1')
axes1.set_ylabel('y_label_axes1')
axes1.set_title('Axes 1 title')

axes2.plot(x,y,'r*-')
axes2.set_xlabel('x_label_axes2')
axes2.set_ylabel('y_label_axes2')
axes2.set_title('Axes 2 title')
```

```python
import pandas as pd
data = [[77,56,34,67,41,56,78,12],
        [58,40,24,38,14,66,54,44],
        [74,32,41,76,13,30,40,45],
        [4,16,66,47,1,52,78,12],
        [14,0,74,28,17,58,52,24]]
df = pd.DataFrame(data , columns=["OOAD","Data Science","Statistics","Python","Sci","Maths","c#","Java"] , index = ["Mahek","Diya","Shweta","Vraj","Het"])
df
```

```python
fig , axes = plt.subplots(2,2, layout="constrained")

axes[0,0].plot(df.index,df.Python,'k--' , label='py')
axes[0,0].set_xlabel('X axis')
axes[0,0].set_ylabel('Y axis')
axes[0,0].set_title('Chart')
axes[0,0].legend()

axes[0,1].plot(df.index,df.Sci,'g--' , label='sci')
axes[0,1].set_xlabel('X axis')
axes[0,1].set_ylabel('Y axis')
axes[0,1].set_title('Chart')
axes[0,1].legend()

axes[1,0].plot(df.index,df.OOAD,'c--' , label='ooad')
axes[1,0].set_xlabel('X axis')
axes[1,0].set_ylabel('Y axis')
axes[1,0].set_title('Chart')
axes[1,0].legend()

axes[1,1].plot(df.index,df.Java,'m--' , label='java')
axes[1,1].set_xlabel('X axis')
axes[1,1].set_ylabel('Y axis')
axes[1,1].set_title('Chart')
axes[1,1].legend()


fig.savefig("filename.png")
```

```python
np.random.seed(3)
x=4+np.random.normal(0,2,24)
y=4+np.random.normal(0,2,len(x))
sizes=np.random.uniform(45,80,len(x))
colors=np.random.uniform(15,95,len(x))
fig , ax = plt.subplots()
ax.scatter(x , y , s=sizes , c=colors)
ax.set(xlim=(0,8) , ylim=(0,8))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.7,cmap='magma',marker='s',edgecolors='red')
plt.colorbar()
```
