```python
import matplotlib.pyplot as plt
import numpy as np 

x1 = np.random.normal(0,50,500)
y1 = np.random.normal(0,50,500)

plt.hist2d(x1, y1, bins=60, cmap='Blues')
plt.colorbar()
```

```python
plt.hexbin(x1, y1, gridsize=10, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(tips['total_bill'])
# Safe to ignore warnings
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=15)
# Safe to ignore warnings
```

```python
sns.distplot(tips['total_bill'])
# Safe to ignore warnings
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

## Pairplot
pairplot will plot pairwise relationships across an entire dataframe 
(for the numerical columns) and supports a color hue argument (for categorical columns) 

```python
sns.pairplot(tips)
```

```python
sns.countplot(data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(tips['total_bill'])
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.boxplot(x='day',y="total_bill", hue='smoker',data=tips,palette='rainbow')
```

```python
sns.boxplot(x='day',y="total_bill",data=tips,palette='rainbow')
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.violinplot(x='day', y="total_bill", data=tips,palette='rainbow')
```

```python
sns.swarmplot(x='day', y="total_bill", hue='smoker',data=tips,palette='cool')
```

```python
sns.swarmplot(x='day', y="total_bill",data=tips,palette='rainbow')
```

```python
sns.pairplot(tips,hue='sex',palette='cool')
#tips.head()
```
